### Random forest Midas regression for GDP: demo

```python
# Midas parameters
# number of periods ahead to predict
h = 1
# number of lags in the low frequency variable
p = 2
# number of lags in the high frequency variables
q = 6
# number of high frequency periods in one low frequency period
m = 3

# random forest parameters
# number of estimators
estimators = 100
# max tree depth
depth = 5
# minimum samples per split
split = 2

# run the model and predict
# declare Random Forest Midas
rfm = Random_Forest_Midas(features, h, p, q, m)
# create regressors
rfm.make_regressors()
# train
rfm.train(estimators, depth, split)
# predict one quarter ahead
y_hat = rfm.predict()
```

### Random forest Midas regression for GDP: predictions

```python
# Midas parameters
# maximum quarter ahead to predict
max_q = 4
# number of lags in the low frequency variable
p = 2
# number of lags in the high frequency variables
q = 6
# number of high frequency periods in one low frequency period
m = 3

# random forest parameters
# number of estimators
estimators = 100
# max tree depth
depth = 5
# minimum samples per split
split = 2

# prediction exercise
# initiate list of predictions
y_hat = np.zeros(max_q)
# loop over forecast periods and predict
for h in range(max_q):
    rfm = Random_Forest_Midas(features, h, p, q, m)
    rfm.make_regressors()
    rfm.train(estimators, depth, split)
    y_hat[h] = rfm.predict()
print(rfm.y_hat)
```

### Random forest VAR for features

```python
# Midas parameters
# number of months ahead to predict
h = 12
# number of lags in the model
p = 2

# random forest parameters
# number of estimators
estimators = 100
# max tree depth
depth = 5
# minimum samples per split
split = 2

# declare random forest VAR
rfv = Random_Forest_Var(features)
# create regressors
rfv.regressors(p)
# train the random forest model
rfv.train(estimators, depth, split)
# predict
rfv.forecast(h)
# plot
rfv.plot_forecast()
```
