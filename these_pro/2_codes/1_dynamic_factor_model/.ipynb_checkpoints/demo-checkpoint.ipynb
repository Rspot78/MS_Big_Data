### Imports

```python
# imports
from feature_processor import *
from dynamic_factor_model import *
```

### Feature processing

```python
# path to data folder
path = '/home/romain/feature_dfm'
# name of file containing monthly features
feature_file = 'features.csv'
# name of file containing the information about monthly data
information_file = 'information.csv'
```

```python
# declare feature processor
fp = FeatureProcessor(path, feature_file, information_file)
# load, process, and plot
fp.data_loader()
fp.data_processor()
fp.plot_processed_data()
```

```python
# overview of processed features in dataframe
features = fp.processed_features
features.head(10)
```

### Dynamic factor model

```python
# declare dynamic factor model
dfm = DynamicFactorModel(feature_dataframe = features, factors = 4, shocks = 3)
# raw factor estimates
dfm.train_raw_factors()
# plot raw factors
dfm.plot_raw_factors()
```

```python
# train parameters
dfm.train_parameters()
# train smoothed factors
dfm.train_smoothed_factors()
# plot smoothed factors
dfm.plot_smoothed_factors()
```

```python
# plot predicted features (in-sample)
dfm.plot_smoothed_features()
```

```python
# train prediction model for GDP
dfm.train_gdp_regression()
# plot predicted GDP (in-sample)
dfm.plot_smoothed_gdp()
```

```python
# predict GDP 8 quarters ahead
dfm.nowcast_gdp(8)
# plot predicted features
dfm.plot_nowcast_features()
```
