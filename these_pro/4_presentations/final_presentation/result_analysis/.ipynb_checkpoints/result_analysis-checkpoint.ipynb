```python
# path to data folder
path = '/home/romain/nowcasting_project/0_nowcasting_documents/5_presentation_brainstorming/result_analysis'
# name of file containing monthly features
feature_file = 'features.csv'
# name of file containing the information about monthly data
information_file = 'information.csv'
# name of file containing gdp data
gdp_file = 'gdp.csv'
# load and process data
dfm = DynamicFactorModel(factors = 4, shocks = 3)
dfm.load(path, feature_file, information_file, gdp_file)
date_list = dfm.date_list
feature_list = dfm.feature_list
number_features = len(feature_list)
```

```python
# indices of prediction months, depending on how many steps ahead are considered
rolling_window_start = '2016-09-30'
max_step_ahead = 4
rolling_window_quarters = 12
rolling_window_start_index = np.where(date_list == rolling_window_start)[0]
prediction_indices = np.zeros((rolling_window_quarters, max_step_ahead))
for step_ahead in range(max_step_ahead):
    first_prediction_index = rolling_window_start_index + 3 * (step_ahead  + 1)
    last_prediction_index = first_prediction_index + 36
    prediction_indices[:, step_ahead] = np.arange(first_prediction_index, last_prediction_index, 3)
```

### RMSE and predictions for the dynamic factor model

```python
# path to data folder
path = '/home/romain/nowcasting_project/0_nowcasting_documents/5_presentation_brainstorming/result_analysis'
# name of file containing monthly features
feature_file = 'features.csv'
# name of file containing the information about monthly data
information_file = 'information.csv'
# name of file containing gdp data
gdp_file = 'gdp.csv'
# number of factors to test
factors = [1, 2, 3, 4, 5, 6]
# number of shocks to test
shocks = [1, 2, 3]
# number of monthly periods in the rolling window
rolling_window = 36
# final sample date of data for optimization (should have known gdp value)
rolling_window_end = '2020m6'
# quarters ahead considered for computing the RMSE when optimizing
h = 1
```

```python
# declare class instance
dfi = DynamicFactorIterator()
# optimize hyperparameters
dfi.hyperparameter_optimizer(path, feature_file, information_file, gdp_file, factors, shocks, rolling_window, rolling_window_end, h)
factors = dfi.optimal_factor
shocks = dfi.optimal_shocks
# compute forecasts and average RMSE with optimal hyperparameters
rolling_window_end = '2019m6'
h = 4
dfi.gdp_prediction_evaluator(path, feature_file, information_file, gdp_file, factors, shocks, rolling_window, rolling_window_end, h)
dfi.feature_prediction_evaluator(path, feature_file, information_file, gdp_file, factors, shocks, rolling_window, rolling_window_end, h)
```

```python
# extract gdp predictions 3, 2, and 1 month before release
dfm_gdp_3 = dfi.storage_prediction_gdp[0::3]
dfm_gdp_2 = dfi.storage_prediction_gdp[1::3]
dfm_gdp_1 = dfi.storage_prediction_gdp[2::3]
# extract gdp rmse
dfm_gdp_rmse = dfi.storage_rmse_gdp
```

```python
# extract feature predictions 3, 2, and 1 month before release
dfm_features_3 = dfi.storage_prediction_features[:, :, 0::3]
dfm_features_2 = dfi.storage_prediction_features[:, :, 1::3]
dfm_features_1 = dfi.storage_prediction_features[:, :, 2::3]
# extract feature rmse
dfm_features_rmse = dfi.mean_rmse_features
dfm_features_rmse_3 = dfi.storage_rmse_features[:, :, 0::3]
dfm_features_rmse_2 = dfi.storage_rmse_features[:, :, 1::3]
dfm_features_rmse_1 = dfi.storage_rmse_features[:, :, 2::3]
```

### RMSE and predictions for the Bayesian VAR model

```python
# load files
path = '/home/romain/nowcasting_project/0_nowcasting_documents/5_presentation_brainstorming/result_analysis'
bvar_gdp = np.zeros((rolling_window_quarters, max_step_ahead))
bvar_features = np.zeros((rolling_window_quarters, number_features, max_step_ahead))
for i in range(4):
    bvar_gdp[:, i] = pd.read_csv(path + '/' + 'bvar_prediction_'+ str(i + 1) + '.csv', delimiter=',', header=None).iloc[:, -1].to_numpy()
    bvar_features[:, :, i] = pd.read_csv(path + '/' + 'bvar_prediction_'+ str(i + 1) + '.csv', delimiter=',', header=None).iloc[:, :-1].to_numpy()
bvar_gdp_rmse = pd.read_csv(path + '/' + 'bvar_rmse.csv', delimiter=',').iloc[:,-1].to_numpy()
bvar_features_rmse = pd.read_csv(path + '/' + 'bvar_rmse.csv', delimiter=',').iloc[:, 1:-1].to_numpy()
```

### RMSE and predictions for the LSTM model

```python
# load files
path = '/home/romain/nowcasting_project/0_nowcasting_documents/5_presentation_brainstorming/result_analysis'
lstm_gdp = np.zeros((rolling_window_quarters, max_step_ahead))
lstm_features = np.zeros((rolling_window_quarters, number_features, max_step_ahead))
for i in range(4):
    lstm_gdp[:, i] = pd.read_csv(path + '/' + 'lstm_prediction_'+ str(i + 1) + '.csv', delimiter=',', header=None).iloc[:, -1].to_numpy()
    lstm_features[:, :, i] = pd.read_csv(path + '/' + 'lstm_prediction_'+ str(i + 1) + '.csv', delimiter=',', header=None).iloc[:, :-1].to_numpy()
lstm_gdp_rmse = pd.read_csv(path + '/' + 'lstm_rmse.csv', delimiter=',').iloc[:,-1].to_numpy()
lstm_features_rmse = pd.read_csv(path + '/' + 'lstm_rmse.csv', delimiter=',').iloc[:, 1:-1].to_numpy()
```

### Plots for GDP nowcasts

```python
# loop over quarters ahead for prediction
fig = plt.figure(figsize = (15, 15))
plt.suptitle('GDP predictions at various horizons', y = 0.93, fontsize = 18, fontweight = 'semibold')
names = ['quarterly_gdp', 'dfm_3', 'bvar', 'lstm']
colours = [(0.1, 0.1, 0.1), (0.8, 0.2, 1), (0.2, 0.6, 0.2), (1.0, 0.6, 0.0)]
labels = ['actual', 'dfm', 'bvar', 'lstm']
for i in range(4):
    date_index = np.array(list(np.arange(1, prediction_indices[0, i])) + list(prediction_indices[:,i]))
    df = dfm.processed_data.iloc[date_index].copy()
    df = df[['quarterly_gdp']]
    df['dfm_3'] = np.nan
    df['dfm_3'].iloc[-12:] = dfm_gdp_3[:, i]
    df['dfm_3'].iloc[-15] = df['quarterly_gdp'].iloc[-15]
    df['bvar'] = np.nan
    df['bvar'].iloc[-12:] = bvar_gdp[:, i]
    df['bvar'].iloc[-15] = df['quarterly_gdp'].iloc[-15]
    df['lstm'] = np.nan
    df['lstm'].iloc[-12:] = lstm_gdp[:, i]
    df['lstm'].iloc[-15] = df['quarterly_gdp'].iloc[-15]    
    plt.subplot(2, 2, i+1)
    plt.title(str(i + 1) + '-step ahead')
    for j in range(4):
        dates = df[[names[j]]].dropna().index
        data = df[[names[j]]].dropna()
        plt.plot(dates, data, linewidth = 2, color = colours[j], label = labels[j])
        if j == 0:
            plot_dates = dates[-16:]
    if i == 0:
        plt.legend()
    plt.xlim(plot_dates[0], plot_dates[-1])
    plt.gca().set_facecolor(color = (0.9, 0.9, 0.9))
    plt.grid(True)
```

### RMSE for GDP nowcasts

```python
# create array of average RMSE
rmse = np.array([list(np.mean(dfm_gdp_rmse, 0)), list(bvar_gdp_rmse), list(lstm_gdp_rmse)])
rmse_dataframe = pd.DataFrame(data = rmse, index=["Dynamic_factor_model", "Bayesian_VAR", "lstm"], columns=["1q_ahead", "2q_ahead", "3q_ahead", "4q_ahead"])
rmse_dataframe
```

### Plots for feature nowcasts

```python
# loop over quarters ahead for prediction
fig = plt.figure(figsize = (20, 35))
plt.suptitle('feature predictions at one-quarter ahead', y = 0.90, fontsize = 18, fontweight = 'semibold')
colours = [(0.1, 0.1, 0.1), (0.8, 0.2, 1), (0.2, 0.6, 0.2), (1.0, 0.6, 0.0)]
labels = ['actual', 'dfm', 'bvar', 'lstm']
date_index = np.array(list(np.arange(1, prediction_indices[0, 0])) + list(prediction_indices[:,0]))
for i in range(number_features):
    names = [feature_list[i], 'dfm_3', 'bvar', 'lstm']
    df = dfm.processed_data.iloc[date_index].copy()
    df = df[[feature_list[i]]]
    df['dfm_3'] = np.nan
    df['dfm_3'].iloc[-12:] = dfm_features_3[0, i, :]
    df['dfm_3'].iloc[-15] = df[feature_list[i]].iloc[-15]
    df['bvar'] = np.nan
    df['bvar'].iloc[-12:] = bvar_features[:, i, 0]
    df['bvar'].iloc[-15] = df[feature_list[i]].iloc[-15]
    df['lstm'] = np.nan
    df['lstm'].iloc[-12:] = lstm_features[:, i, 0]
    df['lstm'].iloc[-15] = df[feature_list[i]].iloc[-15]    
    plt.subplot(8, 4, i+1)
    plt.title(feature_list[i])
    for j in range(4):
        dates = df[[names[j]]].dropna().index
        data = df[[names[j]]].dropna()
        plt.plot(dates, data, linewidth = 2, color = colours[j], label = labels[j])
        if j == 0:
            plot_dates = dates[-20:]
    if i == 0:
        plt.legend()
    plt.xlim(plot_dates[0], plot_dates[-1])
    plt.gca().set_facecolor(color = (0.9, 0.9, 0.9))
    plt.grid(True)
```
