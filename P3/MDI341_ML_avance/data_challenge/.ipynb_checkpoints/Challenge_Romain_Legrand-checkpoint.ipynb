## Challenge description

**Anomaly detection** (or **outlier detection**) comprises the **machine learning** methods aimed at identification of observations that exhibit suspicious behaviour and are very likely to cause a problem. In the **unsupervised learning** framework, no label indicating whether a training observation is anomalous or not is available. Hence, anomalies should be identified in an automatic way by learning the *normal* behavior, that of the vast majority of the observations, and considering those differing significantly from it as *abnormal*. Logically, anomalies are rare in the data and thus fall in *low density* regions: anomaly detection thus boils down to identifying the *tail* of the distribution.

With the ubiquitous deployment of sensors monitoring nearly continuously the health of complex infrastructures, **anomaly detection** can now rely on measurements sampled at a very high frequency, providing a very rich representation of the phenomenon under surveillance. In order to exploit fully the collected information, the observations cannot be treated as multivariate data anymore and a functional analysis approach is required.

The data set is provided by the Airbus and consistst of the measures of the accelerometer of helicopters during  1 minute at frequency 1024 Hertz, which yields time series measured at in total 60 * 1024 = 61440 equidistant time points.

The training set consists of one file, **airbus_train.csv**.

File **airbus_train.csv** contains one observation per row, each observation having 61440 entries, measures with equivalent time distance of 1 / 1024 seconds.

There are in total 1677 training observations.

The training set consists of one file, **airbus_test.csv**, which has the same structure as file **airbus_train.csv**.

There are in total 2511 test observations.

You should submit a file that contains in each row anomaly score for the observation in the corresponding row of the file **airbus_test.csv**. For a sample submission please see the codes below. Please note, that your score should provide ordering which allows to identify anomalies, i.e. the higher the value of the score, the **more abnormal** the observation should be considered.

The performance criterion is the **Area Under the Receiver Operating Characteristic** (AUC), see also:
https://en.wikipedia.org/wiki/Receiver_operating_characteristic#Area_under_the_curve

# I. Dataset exploration

The preliminary exploration of the dataset is crucial, as no preliminary information is available for the dataset. All what is known is that the challenge is based on a **functional dataset** of helicopter accelerometers, and that the objective is to **detect anomalies**.

Because functional data has a natural **time structure**, it is essential to account for this structure so that traditional multivariate methods are not directly applicable. Also, the time strucutre of functional data conditions the nature of the anomalies that can be found in such datasets.

## I.I. Moment analysis

Hubert et al. (2015) identify four types of anomalies commonly encountered in functional datasets.
* **Shift outliers** have the same shape as the majority of observations, but are moved away. They are thus related to a shift in mean or first moment. 
* **Amplitude outliers** also follow the same shape as the majority, but differ in their scale or amplitude. They are thus related to the variance or second moment. 
* **Isolated outliers** exhibit outlying behaviours during a short time interval. This leads to a small number of extreme values, and thus skewed distributions. They are thus related to the skewness or third moment. 
* **Shape outliers** differ in the overall shape of the curve, which may also affect the shape of the distribution and its moments.

Overall, this suggests that unusual or extreme moment values can be related to anomalies in functional data. For this reason, it is worth investigating the moments of the dataset.

```python
# create and fill moment array for train
train_moments = np.zeros((train_obs, 4))
train_moments[:,0] = np.mean(X_train, axis=1)
train_moments[:,1] = np.std(X_train, axis=1)
train_moments[:,2] = skew(X_train, axis=1)
train_moments[:,3] = kurtosis(X_train, axis=1)
titles = ["mean", "standard deviation", "skewness", "kurtosis"]

# plot the histograms for the distribution of moments
fig1 = plt.figure(1, figsize=(20, 10))
plt.suptitle("Histogram of moments distributions: X_train", y=.96, fontsize=20)
for index in range(1,5):
    plt.subplot(2,2,index)
    plt.hist(train_moments[:, index-1], bins = 100, density = True, color = (0, 0.7, 1), edgecolor=(0.4, 0.4, 0.4))
    plt.title(titles[index-1])
plt.show()
    
# convert to pandas dataframe and display descriptive statistics
train_moments_df = pd.DataFrame(data=train_moments, columns=["mean", "st_dev", "skewness", "kurtosis"])
train_moments_stats=round(train_moments_df.describe(percentiles=[.01, .05, .25, .5, .75, .95, .99]),2)
display(train_moments_stats)
```

The descriptive statistics reveal the existence of extreme moment values. 

* the mean exhibits some extreme positive values. Its average is 0.01, close to 0, but 1% of the values are greater than 0.65, more than 5 standard deviations away from the center of the distribution. The maximum value is 1.03, almost ten standard deviations away from the distribution center.

* the volatility shows the same characteristics. Its average is 0.65, but there exist some very small values (0.01 for the minimum), and also a few very large values (max at 6.81, ten standard deviations away from the center).

* skewness is overall equal to 0, meaning the observations are typically centered around 0. However, 1% of the observations have a skewness below -0.66, and 1% above 0.68 (min and max at -3.69 and 1.18 respectively), indicating some asymmetry in the data distribution and thus the presence of extreme values.

* kurtosis displays some extreme positive values. The max is at 22.10, 15 standard deviations away from the center of the distribution.

To summarize, the descriptive statistics and the histograms confirm that some observations have very extreme moments, which may represent good indicators of anomalies.

## I.II. Graphical inspection of anomalies

It is now useful to look at the data graphically to confirm the conclusions of the moment analysis. To do so, I do two things. 

* first, I collect a rough sample of eight "normal" and "outlying" observations. To do this, I use the simple procedure proposed in the starting kit of the challenge: PCA followed by anomaly scores. I then consider the eight most "normal" observations (that with the lowest scores) and the eight most "outlying" observations (that with the highest score). Even though the procedure is probably not extremely efficient, it is plausible that the eight highest values are indeed anomalies, and that the eight lowest are indeed normal.

* second, I collect the eight observations with the most extreme (minimum and maximum) moment values: mean, standard deviation, skewness, kurtosis. 

I finally plot the graphs of these 24 observations. For each observation, I plot the full series (on the left), and the first 200 values (on the right, in order to get a better insight of the local behaviour).

```python
# run pca on train dataset
pca = PCA(n_components = 10, whiten = True)
pca.fit(X_train)
X_train_pca = pca.fit_transform(X_train)
# obtain anomaly score
lof = LocalOutlierFactor(n_neighbors = 5 ,contamination = 'auto', novelty = True)
lof.fit(X_train_pca)
sscore = -lof.score_samples(X_train_pca)
# get index of eight lowest anomaly scores
lowest_scores = (-sscore).argsort()[-8:]
lowest_obs = X_train[lowest_scores,:]
# get index of eight largest anomaly scores
highest_scores = (-sscore).argsort()[:8]
highest_obs = X_train[highest_scores,:]

# get obs with min and max moments
moment_obs = np.zeros([8, train_periods])
moment_obs[0,:] = X_train[np.argmin(train_moments[:,0]),:]
moment_obs[1,:] = X_train[np.argmax(train_moments[:,0]),:]
moment_obs[2,:] = X_train[np.argmin(train_moments[:,1]),:]
moment_obs[3,:] = X_train[np.argmax(train_moments[:,1]),:]
moment_obs[4,:] = X_train[np.argmin(train_moments[:,2]),:]
moment_obs[5,:] = X_train[np.argmax(train_moments[:,2]),:]
moment_obs[6,:] = X_train[np.argmin(train_moments[:,3]),:]
moment_obs[7,:] = X_train[np.argmax(train_moments[:,3]),:]
```

```python
# plots
title_1 = ["min", "max", "min", "max", "min", "max", "min", "max"]
title_2 = ["mean", "mean", "standard deviation", "standard deviation", "skewness", "skewness", "kurtosis", "kurtosis"]
           
# observations that are most normal
fig2 = plt.figure(1, figsize=(15, 30))
plt.suptitle("normal observations", y=.91, fontsize=20)
for index in range(8):
    x1 = lowest_obs[index,:]
    x2 = lowest_obs[index,:200]
    plt.subplot(8,2,2*index+1)
    plt.plot(x1, linewidth=0.1, c=(0.2, 0, 1))
    plt.xlim(0, train_periods)
    plt.title("observation " + str(lowest_scores[index]))
    plt.subplot(8,2,2*index+2)
    plt.plot(x2, linewidth=2, c=(0.2, 0, 1))
    plt.xlim(0, 100)
    plt.title("observation " + str(lowest_scores[index]))
plt.show()
```

```python
# observations most likely to be outliers
fig3 = plt.figure(1, figsize=(15, 30))
plt.suptitle("likely outlier observations", y=.91, fontsize=20)
for index in range(8):
    x1 = highest_obs[index,:]
    x2 = highest_obs[index,:200]
    plt.subplot(8,2,2*index+1)
    plt.plot(x1, linewidth=0.1, c=(0.2, 0, 1))
    plt.xlim(0, train_periods)
    plt.title("observation " + str(highest_scores[index]))
    plt.subplot(8,2,2*index+2)
    plt.plot(x2, linewidth=2, c=(0.2, 0, 1))
    plt.xlim(0, 100)
    plt.title("observation " + str(highest_scores[index]))
plt.show()
```

```python
# observations with extreme moment values
fig4 = plt.figure(1, figsize=(15, 30))
plt.suptitle("extreme moment observations", y=.91, fontsize=20)
for index in range(8):
    x1 = moment_obs[index,:]
    x2 = moment_obs[index,:200]
    plt.subplot(8,2,2*index+1)
    plt.plot(x1, linewidth=0.1, c=(0.2, 0, 1))
    plt.xlim(0, train_periods)
    plt.title(title_1[index] + " " + title_2[index] + " observation")
    plt.subplot(8,2,2*index+2)
    plt.plot(x2, linewidth=2, c=(0.2, 0, 1))
    plt.xlim(0, 100)
    plt.title(title_1[index] + " " + title_2[index] + " observation")
plt.show()
```

The conclusions are the following:
* most observations look stationary around a 0 mean. Many of them exhibit a cyclical behaviour, but for others the cyclicality looks quite loose.
* the simple PCA procedure does not seem to do great overall. Observation 1040 for instance is found to be normal, while it looks in fact suspicious, combining amplitude and shape outlying behaviour.
* the main limit of the PCA procedure is that its outlier identification seems too simplistic. Apparently, it associates 'normal' observations with 'normal' standard deviation (somewhere below 1, the average sample standard deviation), and 'outlier observations' with large standard deviations (above 5). However, outliers can be more complex than just observations with high volatility.
* observations with extreme moment values do exhibit shift, amplitude, isolated and shape outying behaviours. This confirms again that outliers are closely related to extreme moments.

# II. Machine learning strategy

## II.I. A tour on the literature

While functional data analysis is a long-established subject, outlier detection for functional data is a relatively recent field. A good covering of the recent literature is provided by Millan-Roures et al. (2018) and Rousseeuw and Hubert (2018).

Most of the existing litterature focuses on the **unsupervised framework**. Indeed, clean functional datasets with labeled anomalies are hardly ever available, making de facto unsupervised analysis the reference setting. Overall there are two main approaches to outlier detection in functional data.

The first approach relies on **dimensionality reduction**. Indeed, observations in functional datasets are typically high-dimensional, making it practically impossible to analyse them adequately. Most of the time, dimensionality reduction is achieved through principal component analysis (pca) or functional principal component analysis (fpca). Hoffman (2007) applies kernel PCA to hand-written digits functional data. Gervini (2009) introduces the concept of robust pca and uses it to detect outliers in internet traffic and children diabete functional data. Hyndman and Shang (2010) and Sawant et al. (2012) then apply these robust PCA procedures for anomaly detection, using the first few principal components of the dataset.

The second approach consists in adopting **functional depth criteria**. This is by far the most widely used procedure. Depths are calculated for all observations, and outliers are identified as observations with depth beyond a certain threshold. Febrero et al. (2007, 2008) constitute preliminary contributions in this direction. Alternative definitions of functional depth have then been developed and applied in the context of outlier detection, including Sun and Genton (2011), Gervini (2012), Arribas-Gil and Romo (2014), Hubert et al. (2015), Kuhnt and Rehage (2016), Nagy et al. (2017), Dai and Genton (2018), and Huang and Sun (2019). Most of the time, these papers try to overcome the limits of traditional depth measures, such as their inefficiency at capturing shape outliers.

Finally, a limited number of contributions attempt at proposing alternative approaches. Fraiman and Svarc (2013) propose an adaptative method based on random projections. Millan-Roures et al. (2018) detect outliers in urban water functional data with functional archetype analysis. Staerman et al. (2019) develop a functional isolation forest algorithm to separate outliers from normal functional data. Dai et al. (2020), finally, propose to apply simple transformations to functional datasets in order to make outliers easier to detect.

While unsupervised outlier identification is the norm, it is worth mentioning one contribution relying on a **semi-supervised** framework. Barreyre et al. (2017) use a standard pca procedure to reduce the dimensionality of their datasets, but note that that using the first few components may not be suitable for anomaly detection. On the contrary, because the anomaly is rare by definition, it is unlikely to be found in the first components which represent a large portion of the variance of the data. They thus aim at identifying the components contributing most to anomaly detection, and to this end compare their distributions with that of the same components for a clean dataset without outliers.



## II.II. Pitfalls in the challenge

Three main difficulties arise in this challenge:

* the data is functional, and labels are not available. Therefore, supervised learning methods are not directly applicable. Also, hyperparameter optimisation is not possible due to the lack of cross-validation data.

* the challenge consists in creating a score-based anomaly ranking. Therefore, a scoring methodology is required. However, a large number of methodologies for the detection of functional anomalies are based on graphical representations such as bagplots and boxplots (Hyndman and Shang (2010), Sun and Genton (2011), Hubert et al. (2015), just to cite a few). This eliminates de facto all these methodologies, restricting by much the set of tools available.

* The challenge explicitly limits the code to one Jupyter Notebook. However, a considerable number of libraries are developed in R (Rainbow, fda.usc, roahd, mrfDepth) and are thus not available.

Fortunately, Python also proposes a number of interesting libraries. The FIF library proposes an implementation of the Functional Isolation Forest algorithm. The scikit-fda package offers a number of functional data analysis algorithms, including distance-based algorithms for outlier detection. And, of course, the popular library Scikit Learn offers a range of algorithms for outlier detection like one-class SVM, IsolationForest, and local outlier factor.

## II.III. Preliminary machine learning approaches

Given the above analysis, a natural strategy consists in using the outlier detection libraries offered by Python. In particular, we can retain the following strategies:

* the local outlier factor of Breunig et al. (2000).
* the functional isolation forest of Staerman et al (2019). 
* the one-class SVM of Scholkopf et al. (2000).
* the directional outlyingness of Dai and Genton (2018).

Because the dataset is of high dimension, the performance of the estimate may be affected. Therefore, for each  methodology, dimension reduction is first applied to the data. Three possibilities are considered:

* principal component analysis with 5 components.
* Fourier transform with 50 components.
* wavelet basis with 25 approximation and detail coefficients (symlets 5 basis).

The latter two transforms are motivated by the time-series nature of the observations.

```python
# matrix of records
auc_scores = np.zeros((3,4))
```

```python
# PCA with 5 components, on train and test
components = 5
pca = PCA(n_components = components, whiten = True)
pca.fit(X_train)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.fit_transform(X_test)
```

```python
# PCA: functional isolation forest
np.random.seed(0)
fif = FIForest(X_train_pca, D="gaussian_wavelets", time=np.arange(1, components+1), innerproduct="auto", alpha=0.5, limit=15)
# get anomaly score on test set
as2 = -fif.compute_paths(X_test_pca)
# save the anomaly scores to file
np.savetxt('2_pca_fif.csv', as2, fmt = '%1.6f', delimiter=',')
# record auc score from leaderboard
auc_scores[0,1] = 0.4424
```

```python
# PCA: one-class SVM
ocsvm = OneClassSVM(gamma='auto')
ocsvm.fit(X_train_pca)
# get anomaly score on test set
as3 = -ocsvm.score_samples(X_test_pca)
# save the anomaly scores to file
np.savetxt('3_pca_ocsvm.csv', as3, fmt = '%1.6f', delimiter=',')
# record auc score from leaderboard
auc_scores[0,2] =  0.7030
```

# III. Data augmentation

## III.I Motivation and methodology

A difficulty with the dataset at hand is that it is rather small (1677 observations in the train set), and that anomalies are unknown. It could therefore be very helpful to have more data at our disposal, and ideally to obtain identified anomalies to run the training.

Based on this, I propose to augment the dataset at hand with artifical observations, in the hope that it can improve performance. Concretely, I create a class (AirbusObservationGenerator, see below), that generates two type of observations.

**Normal observations** are generated cheaply, in the following way:
* draw n series randomly from the train set, with n small (e.g. n=5) and average them to get a new base series. 
* add a tiny noise to the base series to generate randomness.
* randomly sample 'normal' values of mean, standard deviation and skewness for the series.
* adjust the base series so that it matches the drawn mean, standard deviation and skewness.


A few remarks on the generation process:

* because functional observations are assumed to be drawn from the same distibution, n has to be small to avoid convergence to the mean of the distribution and generate similar base series.
* the noise is defined as Gaussian with 0 mean and standard deviation being 1/200 of that of the base series.
* The 'normal' moments are defined as the central 90% of the empirical distribution over the train set, drawing being done from a Gaussian kernel density estimation.
* adjustement of the series is done sequentially: first the skewness is rectified using the sinh-arcsinh distributions of Jones and Pewsey (2009); then the standard deviation is corrected by scaling the series, which leaves skewness unaffected; finally, the mean is adjusted by shifting, which does not affect the previous two moments.

This provides a cheap, unlimitted supply of different normal observations. To create **outlier observations**, a normal observation is first generated, then it is modified to create an outlier. Following the taxonomy of Hubert et al. (2015), four types of outliers can be generated:

* shape outliers are generated by shifting a fraction gamma of the observation to a non-normal level. Given the distirbution of the moments in the train set (see section I.I.), I define a non-normal mean as e mean being 6 to 8 standard deviations away from the average.

* amplitude outliers are generated by increasing the volatility of a fraction gamma of the observation to a non normal level. I define non normal volatility either as an excessively small level of variance (0.1 to 1 time the minimum variance of the train set), or excessively large level (1 to 2 times the maximum variance of the train set)

* shape outliers are generated by creating a linear deviation over a fraction gamma of the observation. The slope is determined randomly over the (-0.25, 0.25) interval, and scaled by the standard deviation of the observation.

These three types of outliers are sufficiently flexible to cover most outliers. For instance, isolated outliers can be generated by implementing a magnitude outlier over a very brief fraction of the observation.

```python
class AirbusObservationGenerator:
    '''a class which generates artificial observations for the airbus dataset;
    generated observations can be normal observations or custom outliers'''
    
    def __init__(self, X=X_train, d=5, alpha = 0.05):
        self.X = X
        self.nobs = self.X.shape[0]
        self.periods = self.X.shape[1]
        self.d = d
        self.alpha = alpha
        # get first three moments of each observation in X
        self.m1 = np.mean(X_train, axis=1)
        self.m1_std = np.std(self.m1)
        self.m2 = np.std(X_train, axis=1)
        self.m2_min = min(self.m2)
        self.m2_max = max(self.m2)
        self.m3 = skew(X_train, axis=1)                              
        # get empirical distributions for non outlier (extreme) values
        self.m1_kde = self.non_outlier_kernel(self.m1)
        self.m2_kde = self.non_outlier_kernel(self.m2)
        self.m3_kde = self.non_outlier_kernel(self.m3)

    def skew_observation(self, x, eps):
        '''modify sample skewness using the sinh-arcsinh distributions of Jones and Pewsey(2009)'''      
        return np.sinh(np.arcsinh(x)-eps)

    def skewness_error(self, eps, x, target):
        '''return the squared difference between target and skewness of x obtained with eps'''
        return (skew(self.skew_observation(x, eps)) - target)**2
    
    def non_outlier_kernel(self, x):
        '''return an empirical pdf of x, excluding the extreme alpha percent of each tail'''
        return gaussian_kde(np.sort(x)[round(self.alpha*self.nobs):round((1-self.alpha)*self.nobs)])
    
    def generate_normal_observation(self):
        '''generate a normal (non outlier) artificial observation for the airbus dataset'''
        # randomly draw mean, std and skewness from the non-outlier distributions
        obs_m1 = self.m1_kde.resample(1)[0][0]
        obs_m2 = self.m2_kde.resample(1)[0][0]
        obs_m3 = self.m3_kde.resample(1)[0][0]
        # sample d observations randomly from X
        obs = self.X[np.random.choice(self.nobs,self.d),:]
        # obtain an average observation (scale by 10 to make sure we can modify skewness)
        obs = np.mean(obs, axis=0)
        obs += np.std(obs) / 200 * standard_normal()
        obs *= 10
        # adjust skewness so it corresponds to obs_m3
        eps = minimize(self.skewness_error, 0, args=(obs, obs_m3)).x[0]
        obs = self.skew_observation(obs,eps)
        # adjust standard deviation so it corresponds to obs_m2 (it does not effect skewness)
        obs_std = np.std(obs)
        obs *= obs_m2 / obs_std
        # adjust mean so it corresponds to obs_m1 (it does not affect skewness nor standard deviation)
        obs_mean = np.mean(obs)
        obs += obs_m1 - obs_mean
        return obs
    
    def generate_normal_observations(self,n):
        '''generate n normal (non outlier) artificial observations for the airbus dataset'''
        obs = np.zeros((n, self.periods))
        for i in range(n):
            obs[i,:] = self.generate_normal_observation()
        return obs
    
    def generate_outlier_observation_t1(self, gamma=0.1, sign=1):
        '''generate an outlier observation of type 1: shift outlier
        outlying behaviour is exerted over a fraction gamma of the observation, with location randomly decided'''
        # generate a normal observation
        obs = self.generate_normal_observation()
        # randomly decide where shift occurs
        start_period, end_period = self.shift_periods(gamma)
        # randomly determine shift size: from 6 to 8 std away from current value
        shift_amplitude = uniform(6, 8) * sign * self.m1_std
        obs[start_period:end_period] += shift_amplitude
        return obs
    
    def generate_outlier_observations_t1(self, n=10, gamma=0.1, sign=1):
        '''generate n outlier observation of type 1: shift outlier'''
        obs = np.zeros((n, self.periods))
        for i in range(n):
            obs[i,:] = self.generate_outlier_observation_t1(gamma)
        return obs
    
    def generate_outlier_observation_t2(self, gamma=0.1, sign=1):
        '''generate an outlier observation of type 2: amplitude outlier
        outlying behaviour is exerted over a fraction gamma of the observation, with location randomly decided'''
        # generate a normal observation
        obs = self.generate_normal_observation()
        # randomly decide where shift occurs
        start_period, end_period = self.shift_periods(gamma)
        obs_std = np.std(obs)
        # amplitude can be shifted up (1 to 2 time sample max)
        if sign == 1:
            obs[start_period:end_period] *= self.m2_max * uniform(1,2) / obs_std
        # amplitude can be shifted down (0.1 to 1 time sample min)
        elif sign == -1:
            obs[start_period:end_period] *= self.m2_min * uniform(0.1,1) / obs_std
        return obs
    
    def generate_outlier_observations_t2(self, n=10, gamma=0.1, sign=1):
        '''generate n outlier observation of type 2: amplitude outlier'''
        obs = np.zeros((n, self.periods))
        for i in range(n):
            obs[i,:] = self.generate_outlier_observation_t2(gamma, sign)
        return obs
    
    def generate_outlier_observation_t3(self, gamma=0.1, sign=1):
        '''generate an outlier observation of type 3: shape outlier
        outlying behaviour is exerted over a fraction gamma of the observation, with location randomly decided'''
        # generate a normal observation
        obs = self.generate_normal_observation()
        obs_std = np.std(obs)
        # randomly decide where shift occurs
        start_period, end_period = self.shift_periods(gamma)
        # create random shape shift (positive or negative)
        shift = np.arange(end_period - start_period) * uniform(0.1,0.25) * sign * obs_std * 0.003
        obs[start_period:end_period] += shift
        return obs
    
    def generate_outlier_observations_t3(self, n=10, gamma=0.1, sign=1):
        '''generate n outlier observation of type 3: shape outlier'''
        obs = np.zeros((n, self.periods))
        for i in range(n):
            obs[i,:] = self.generate_outlier_observation_t3(gamma, sign)
        return obs
    
    def shift_periods(self, gamma):
        if gamma < 1:
            start_period = randint(0, math.floor((1-gamma)*self.periods))
            end_period = start_period + math.floor((gamma)*self.periods)
        else:
            start_period, end_period = 0, self.periods
        return start_period, end_period
```

## III.III Supervised learning with data augmentation

The main advantage of the data generating process is that it provides labeled observations. It is thus possible to use this labelled information to conduct supervised learning procedures. 

I will focus on two state-of-the-art classification algorithms: neural networks, and XGBoost. Because these algorithms can be biased towards the majority class, I generate balanced datasets for normal and outlier observations. Precisely, I generate 2000 normal observations, and 1600 anomalies. I then train the models on this dataset.

After a few attempts, the best neural network is a simple setting with four dense layers, 70 neurons for each layer, a relu activation, batch normalisation (batch of size 200), and 20% dropout rate.

For the XGBoost algorithm, I adopt a standard setting with a maximum depth of 8 for the trees, and a learning of 0.3. Becuase the classes are balanced in the dataset, I set a weight of 0.5 for each class.

```python
# generate simulated dataset
aog = AirbusObservationGenerator(X_train, 5, 0.05)
# normal observations
X_train_gen = aog.generate_normal_observations(2000)
# outlier observations
X_train_gen = np.concatenate((X_train_gen,aog.generate_outlier_observations_t1(250,1,1)),axis=0)
X_train_gen = np.concatenate((X_train_gen,aog.generate_outlier_observations_t1(250,1,-1)),axis=0)
X_train_gen = np.concatenate((X_train_gen,aog.generate_outlier_observations_t2(250,1,1)),axis=0)
X_train_gen = np.concatenate((X_train_gen,aog.generate_outlier_observations_t2(250,1,-1)),axis=0)
X_train_gen = np.concatenate((X_train_gen,aog.generate_outlier_observations_t3(250,1,1)),axis=0)
X_train_gen = np.concatenate((X_train_gen,aog.generate_outlier_observations_t3(250,1,-1)),axis=0)
X_train_gen = np.concatenate((X_train_gen,aog.generate_outlier_observations_t2(100,0.0005,1)),axis=0)
# labels
Y_train_gen = np.concatenate((np.zeros((2000)),np.ones((1600))),axis=0)
# shuffle
shuffle_index = np.arange(X_train_gen.shape[0])
np.random.shuffle(shuffle_index)
X_train_gen = X_train_gen[shuffle_index]
Y_train_gen = Y_train_gen[shuffle_index]
```

```python
# deep learning

# architecture of the network
nn_epochs = 10
nn_batch_size = 200
nn_verbose = 1
nn_input_size = train_periods
nn_layer_size = 70
nn_validation_split = 0.1
nn_dropout = 0.2

# building of the model
model = tf.keras.models.Sequential()

# dense layer 1
model.add(keras.layers.Dense(units=nn_layer_size,input_shape=(nn_input_size,), name='dense_layer_1'))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Activation("relu"))

# dense layer 2 with dropout
model.add(keras.layers.Dense(nn_layer_size, name='dense_layer_2'))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Activation("relu"))
model.add(keras.layers.Dropout(nn_dropout))

# dense layer 3 with dropout
model.add(keras.layers.Dense(nn_layer_size, name='dense_layer_3'))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Activation("relu"))

# dense layer 4 with dropout
model.add(keras.layers.Dense(nn_layer_size, name='dense_layer_4'))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Activation("relu"))
model.add(keras.layers.Dropout(nn_dropout))

# decision layer
model.add(keras.layers.Dense(1, name='dense_layer_final', activation='sigmoid'))

# summary of the model
model.summary()
```

```python
# compilation
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# training of the model
model.fit(X_train_gen, Y_train_gen, batch_size=nn_batch_size, epochs=nn_epochs, verbose=nn_verbose, \
         validation_split=nn_validation_split);

# save model
model.save('nn_model.h5')
```

```python
# load model if needed
# model = tf.keras.models.load_model('nn_model.h5')

# obtain anomaly score on test set
as21 = model.predict(X_test)

# save the anomaly scores to file
np.savetxt('21_nn.csv', as21, fmt = '%1.6f', delimiter=',')

# record auc score from leaderboard
print("AUC on leaderboard is " + str(0.7519))
```

```python
# XGBoost model

# declare XGBoost model
xgb = XGBClassifier(scale_pos_weight=0.5, max_depth=8, eta=0.3);

# fit model to data
xgb.fit(X_train_gen, Y_train_gen);

# save model to file
pickle.dump(xgb, open("xgb.pickle.dat", "wb"))

# obtain predictions on the test set
as22 = xgb.predict_proba(X_test)[:,1]

# save the predictions
np.savetxt('22_xgb.csv', as22, fmt = '%1.6f', delimiter=',')

# record auc score from leaderboard
print("AUC on leaderboard is " + str(0.6716))
```

By far, the deep learning model is the one that performs best, with an AUC of 0.7519. The XGboost model is fairly disappointing, with an AUC of 0.6716 only.

# Conclusion

This challenge proposed to visit a set of unlabelled functional data. After analysing the properties of the dataset et reveiwing the litterature, I followed two main approaches.

The first consisted in reducing the dimension of the dataset by the way of projection on different bases (PCA, Fourier, wavelets). A number of methodologies were then applied to the reduced dataset to identify outliers: local outlier factor, functional isolation forest, one-class SVM and outlyingness depth. These approaches yielded fair performance, with an AUC around 0.72 for the best performance.

The second approach consisted in augmenting the dataset with simulated data. This approach permitted to generate both normal and outlier observations. An unsupervised setting on the generated data did not improve the performance. A supervised approach exploiting the label information however did manage to improve the results. A simple neural network trained on a labeled dataset distributed at par between normal and outlier observations managed to achieve an AUC around 0.75. An XGBoost model on the other hand yielded poor performance.
