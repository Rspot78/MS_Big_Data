```python
df2 = pd.DataFrame({
    'a': [1, 2, 3, 4, 5, 6, 7],
    'b': [1, 4, 9, 16, 25, 36, 49],
})
```

```python
# Custom axes ticks / labels:
ax = df2.plot(x='a', y='b', label='squares')

ax.set_xticks([1, 3, 5, 7])
ax.set_yticks(df2.b)
ax.set_xticklabels(['one', 'three', 'five', 'seven'], rotation=30)
```

```python
# Multiple plots on 1 figure

fig, axes = plt.subplots(2, 2, sharex=True)

axes[0, 0].plot(df.index, df.x, c='green')

axes[0, 1].plot(df.index, df.y, c='purple')

axes[1, 0].plot(df.index, df.x, c='green', label='x')
axes[1, 0].plot(df.index, df.z, c='orange', label='z')
axes[1, 0].legend()

cmap = colors.ListedColormap(['green', 'purple', 'orange'])
df.plot(ax=axes[1, 1], colormap=cmap)

plt.subplots_adjust(hspace=0.05)
```
